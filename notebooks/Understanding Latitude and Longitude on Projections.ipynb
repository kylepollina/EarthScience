```python
import matplotlib.pyplot as plt
from cartopy import crs as ccrs
```

```python
ccrs.PlateCarree()
```

```python
lat_range = [lat for lat in range(-90, 90, 10)]
lon_range = [lon for lon in range(-180, 180, 10)]

lats = []
lons = []

for x in lon_range:
    for y in lat_range:
        lats.append(y)
        lons.append(x)
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(1,1,1, projection=ccrs.PlateCarree())
ax.gridlines(draw_labels=True)
ax.coastlines()
ax.scatter(x=lons, y=lats)
```

```python
proj = ccrs.Robinson()
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(1,1,1, projection=proj)
ax.gridlines(draw_labels=True)
ax.coastlines()
ax.scatter(
    x=[proj.transform_point(x, 0, ccrs.PlateCarree())[0] for x in lons],
    y=[proj.transform_point(0, y, ccrs.PlateCarree())[1] for y in lats]
)
```

```python
proj = ccrs.AlbersEqualArea()
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(1,1,1, projection=proj)
ax.coastlines()
ax.gridlines(draw_labels=True)
ax.scatter(
    x=[proj.transform_point(x, 0, ccrs.PlateCarree())[0] for x in lons],
    y=[proj.transform_point(0, y, ccrs.PlateCarree())[1] for y in lats]
)
```
